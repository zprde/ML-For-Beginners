> import related packages

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model, model_selection
```

> import the datasets from sklearn lib as X,y <br/>

X should be: 
- ages: age in years
- sex
- bmi: body mass index
- bp average blood pressure
- s1 tc, total serum cholesterol
- s2 ldl, low-density lipoproteins
- s3 hdl, high-density lipoproteins
- s4 tch, total cholesterol / HDL
- s5 ltg, possibly log of serum triglycerides level
- s6 glu, blood sugar level 

y should be a quantitative measure of disease progression one year after baseline

```python
X, y = datasets.load_diabetes(return_X_y=True)
for index, feature in enumerate(X[:9]):
    print(f'{feature} - {y[index]}\n')
X_bmi = X[:, 2].reshape((-1, 1))
```

> Split the data to train data and test data

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(X_bmi, y, test_size=0.66)
```

> Train the model

```python
model = linear_model.LinearRegression()
model.fit(X_train, y_train)
```

> Test the model with test data

```python
y_predict = model.predict(X_test)
```

> Compare test data with predicated data

```python
plt.scatter(X_test, y_test, color='black')
plt.plot(X_test, y_predict, color='blue', linewidth=3)
plt.xlabel('Scaled BMIs')
plt.ylabel('Disease Progression')
plt.title('A Graph Plot Showing Diabetes Progression Against BMI')
plt.show()
```
